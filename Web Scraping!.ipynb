```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
website = 'https://data.bls.gov/cgi-bin/surveymost?bls'
```

```python
Unemployment_Rate = requests.post('https://data.bls.gov/cgi-bin/surveymost',data = {"series_id":"LNS14000000","survey":"bls","format":"","html_tables":"","delimiter":"","catalog":"","print_line_length":"","lines_per_page":"","row_stub_key":"","year":"","date":"","net_change_start":"","net_change_end":"","percent_change_start":"","percent_change_end":""})
```

```python
soup_Unemployment_Rate = BeautifulSoup(Unemployment_Rate.content, 'html.parser')
soup_Unemployment_Rate
```

```python
find_table = soup_Unemployment_Rate.find("table", attrs={'class':'regular-data'})
headers = find_table.thead.find("th")
find_table = soup_Unemployment_Rate.find("table", attrs={'class':'regular-data'})
row_header = find_table.tbody.find_all("tr")
```

```python
values = {}
for item in row_header:
    values[item.find("th").get_text()] = list(map(lambda x:x.get_text(), item.find_all("td")))
print(values)
row_hea =[]# ['Year']
for item in row_header:
    row_hea.append(item.find("th").get_text())
print(row_hea)
```

```python
df = pd.DataFrame(values)
df.index = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
df_Unemployment_Rate  = df[df.columns[-5:]]
```

```python
df_Unemployment_Rate.reset_index(inplace=True)
df_Unemployment_Rate = pd.melt(df_Unemployment_Rate, id_vars=['index'], value_vars=list(df_Unemployment_Rate.columns[2:]))
df_Unemployment_Rate.rename(columns={'value':'Unemployment_Rate'}, inplace=True)
```

```python
df_Unemployment_Rate["Date"] = df_Unemployment_Rate["index"] + "-" + df_Unemployment_Rate["variable"]
df_Unemployment_Rate.drop(['index','variable'], axis='columns', inplace=True)
#df_Unemployment_Rate
```

```python
Oil_Price = requests.get('https://www.eia.gov/dnav/pet/hist/LeafHandler.ashx?n=pet&s=f000000__3&f=m')
```

```python
soup_Oil_Price = BeautifulSoup(Oil_Price.content, 'html.parser')
soup_Oil_Price
```

```python
find_table = soup_Oil_Price.find("table", attrs={'class':'FloatTitle'})
headers = find_table.thead.tr.find_all("th")
find_table = soup_Oil_Price.find("table", attrs={'class':'FloatTitle'})
row_header = find_table.tbody.find_all("tr")
```

```python
row_header
```

```python
values = {}
for item in row_header:
    td_row = item.find_all("td")
    if len(td_row)>1 :
        values[item.find("td").get_text().strip()] = list(map(lambda x:x.get_text(), td_row[1:]))
print(values)
```

```python
df = pd.DataFrame(values)
df.index = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
df_Oil_Price  = df[df.columns[-5:]]
```

```python
df_Oil_Price.reset_index(inplace=True)
df_Oil_Price = pd.melt(df_Oil_Price, id_vars=['index'], value_vars=list(df_Oil_Price.columns[2:]))
df_Oil_Price.rename(columns={'value':'Oil_Price'}, inplace=True)
```

```python
df_Oil_Price["Date"] = df_Oil_Price["index"] + "-" + df_Oil_Price["variable"]
df_Oil_Price.drop(['index','variable'], axis='columns', inplace=True)
#df_Oil_Price
```

```python
final_df = pd.merge(df_Unemployment_Rate, df_Oil_Price, on= ["Date"])
final_df = final_df[['Date', 'Unemployment_Rate','Oil_Price']]
```
